```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import os
import time
```

```python
def pull(start,end):
    container = []
    string = 'Bitcoin since:'+start+' until:'+end
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(string).get_items()):
        if i > 1500:
            break
        container.append([tweet.user.username, tweet.date, tweet.likeCount, 
                                 tweet.sourceLabel, tweet.content, tweet.retweetCount, tweet.replyCount,
                         tweet.user.location,tweet.user.followersCount,tweet.user.friendsCount,
                         tweet.user.verified])
        
    tweets_df = pd.DataFrame(container, columns=["User", "Date Created", "Number of Likes", "Source of Tweet", "Tweet",
                                                       "Retweet counts", "Reply counts","location",
                                                "followers","friends","verified"])
    filename = 'Bitcoin'+start+'.csv'
    tweets_df.to_csv(filename)
```

```python
loops = pd.read_excel('date_loop.xlsx', sheet_name="Sheet1", dtype = str)
loops.head(5)
```

```python
loops.shape
```

```python
t1 = time.time()
for i in range(0,len(loops)):
    start = str(loops.loc[i,"Start_Date"])[0:11]
    end = str(loops.loc[i,"End_Date"])[0:11]
    pull(start,end)
    print(start,end)

t2 = time.time()

print(t2-t1)
```
